```
div_by_7_and_5 = [x  for x in range(1000,2001) if x%5 == 0 and x%7 == 0]
print(div_by_7_and_5)
```

```

from math import sqrt
perfect_squares= [x for x in range(500,1001) if sqrt(x) == int(sqrt(x)) and x%5 == 0]
print(perfect_squares)
```
